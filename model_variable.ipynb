```python
import pandas as pd
data=pd.read_csv('CIBERSORTx_GSE76772_clean.csv')
samples=data['Mixture'].values
samples
```

```python
data.set_index('Mixture',inplace=True)
data
```

```python
genes=pd.read_csv('HUGO_GSE76772.csv')
genes.set_index('Gene_Symbol',inplace=True)
genes=genes[samples]
genes=genes.T
genes.head()
```

```python
data['IL12']=genes['IL12A']+genes['IL12B']
data['HMGB1']=genes['HMGB1']
data['IL6']=genes['IL6']+genes['IL6ST']+genes['IL6RA']
data['IL10']=genes['IL10']+genes['IL10RB']+genes['IL10RA']
data['IFNG']=genes['IFNG']
data.head()

```

```python
SS_variable_list=['Tn', 'Th', 'Tc', 'Tr', 'Dn', 'D', 'Mn', 'M', 'C', 'N', 'A','Ig','H','IL12','IL10','IL6']
IC_variable_list=['Tn', 'Th', 'Tc', 'Tr', 'Dn', 'D', 'Mn', 'M', 'C', 'N', 'A','Ig','H','IL12','IL10','IL6']
Immune_cells=['Tn', 'Th', 'Tc', 'Tr', 'Dn', 'D', 'Mn', 'M']
model_cells=['Tn', 'Th', 'Tc', 'Tr', 'Dn', 'D', 'M','Mn']
```

```python
index=data.index
# rearrange the data corresponding to variables
full_data=pd.DataFrame(index=index)
full_data['Sample ID']=data.index
full_data['Tn']=data[['T cells CD4 naive', 'T cells CD4 memory resting', 'NK cells resting']].sum(axis=1)
full_data['Th']=data['T cells CD4 memory activated']+data['T cells follicular helper']
full_data['Tc']=data['T cells CD8']+data['NK cells activated']
full_data['Tr']=data['T cells regulatory (Tregs)']
full_data['Dn']=data['Dendritic cells resting']
full_data['D']=data['Dendritic cells activated']
full_data['M']=data['Macrophages M1']+data['Macrophages M2']
full_data['Mn']=data[['Monocytes', 'Macrophages M0']].sum(axis=1)
full_data['C']=""
full_data['N']=""
full_data['IL12']=data['IL12']
full_data['IL10']=data['IL10']
full_data['IL6']=data['IL6']
full_data['H']=data['HMGB1']
full_data['Ig']=data['IFNG']
#full_data['Tumor_weight']=data['Tumor_weight']
#full_data['dim']=data['Tumor_Size']

full_data['TIC']=full_data[['Tn', 'Th', 'Tc', 'Tr','Dn', 'D', 'M','Mn']].sum(axis=1)

#Substitute zeros with something negligibly small, but positive
mincellvalue=full_data[Immune_cells].to_numpy()[:]
mincellvalue=mincellvalue[mincellvalue>0].min()
for key in Immune_cells:
    full_data.loc[full_data[key]<=0, key]=0.1*mincellvalue
```

```python
import numpy as np

meandim=np.empty(3)
#meandim[0]=full_data_meta['dim'].mean()
for i in range(3):
    meandim[i]=full_data[full_data['samples']==i+1]['tumor_size'].mean()


meandim
```

```python
# Mean total immune cell count global and for each cluster
meanTIC=np.empty(3)

for i in range(3):
    meanTIC[i]=full_data[full_data['samples']==i+1]['TIC'].mean()
meanTIC
```

```python
s1=full_data[full_data.samples==1]
s2=full_data[full_data.samples==2]
s3=full_data[full_data.samples==3]
```

```python
#total cell number TCN 
#collaborators suggested C:I:N ratio 0.955:0.04:0.005
# Evaluating number of cancer and necrotic cells from tumor size and immune cells
#We subtract the total number of immune cells from total number of cells to get the total number of cancer+necrosis. Then based on 0.955:0.005 ratio when multiplied by 955/960 we only get cancers.
#Necrotic cells are 1/191 of cancer cells according to 0.955:0.005 ratio
#literature says 4.5e1 is the density of epithelial cells. So by this multiplication we will make sure the average density of immune cells matches this number
#literature says 4.5e1 is the density of epithelial cells. So by this multiplication we will make sure the average density of cancer cells and necrotic cells matches this number. The reason we use 0.955 for both is because before this multiplication the ratio 0.955:0.005 holds between C and N. So we devide by 0.955 for both to keep the ratio.

s1.C=955*((s1.tumor_size/meandim[0])-0.04*(s1.TIC/meanTIC[0]))/960 
s1.N=s1.C/191               
s1[Immune_cells]*=4.5e1/meanTIC[0]      
s1[['C', 'N']]*=4.5e1/0.955        
s1['A']=s1[['Tn', 'Th', 'Tc', 'Tr', 'Dn', 'D', 'Mn', 'M']].sum(axis=1)*2
s1
```
